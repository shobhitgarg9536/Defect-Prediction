```python
import pandas as pd
```

```python
ml_df = pd.read_csv("apache.csv")
zxing_df = pd.read_csv("ar3.csv")
ml_df = ml_df[['AvgLine','CountLine','CountLineBlank','CountLineCode',
             'CountLineCodeDecl','CountLineCodeExe','CountLineComment','CountSemicolon',
             'CountStmt','CountStmtDecl','CountStmtExe','SumCyclomatic',
             'SumCyclomaticModified','SumCyclomaticStrict','SumEssential','isDefective']]

zxing_df = zxing_df[['halstead_effort','halstead_length','decision_count','total_operands','executable_loc','branch_count',
                   'halstead_vocabulary','defects']]

#print(ml_df.info())
new_dataset = pd.DataFrame()
new_dataset['AvgLine'] = zxing_df['halstead_effort']
new_dataset['CountLine'] = zxing_df['halstead_length']
new_dataset['CountLineBlank'] = zxing_df['decision_count']
new_dataset['CountLineCode'] = zxing_df['total_operands']
new_dataset['CountLineCodeDecl'] = zxing_df['executable_loc']
new_dataset['CountLineCodeExe'] = zxing_df['halstead_effort']
new_dataset['CountLineComment'] = zxing_df['halstead_length']
new_dataset['CountSemicolon'] = zxing_df['total_operands']
new_dataset['CountStmt'] = zxing_df['halstead_effort']
new_dataset['CountStmtDecl'] = zxing_df['decision_count']
new_dataset['CountStmtExe'] = zxing_df['branch_count']
new_dataset['SumCyclomatic'] = zxing_df['halstead_effort']
new_dataset['SumCyclomaticModified'] = zxing_df['halstead_vocabulary']
new_dataset['SumCyclomaticStrict'] = zxing_df['halstead_vocabulary']
new_dataset['SumEssential'] = zxing_df['halstead_vocabulary']
new_dataset['defects'] = zxing_df['defects'].map({False:0,True:1})
print(new_dataset.info())

new_ml = pd.DataFrame()
new_ml['AvgLine'] = ml_df['AvgLine']
new_ml['CountLine'] = ml_df['CountLine']
new_ml['CountLineBlank'] = ml_df['CountLineBlank']
new_ml['CountLineCode'] = ml_df['CountLineCode']
new_ml['CountLineCodeDecl'] = ml_df['CountLineCodeDecl']
new_ml['CountLineCodeExe'] = ml_df['CountLineCodeExe']
new_ml['CountLineComment'] = ml_df['CountLineComment']
new_ml['CountSemicolon'] = ml_df['CountSemicolon']
new_ml['CountStmt'] = ml_df['CountStmt']
new_ml['CountStmtDecl'] = ml_df['CountStmtDecl']
new_ml['CountStmtExe'] = ml_df['CountStmtExe']
new_ml['SumCyclomatic'] = ml_df['SumCyclomatic']
new_ml['SumCyclomaticModified'] = ml_df['SumCyclomaticModified']
new_ml['SumCyclomaticStrict'] = ml_df['SumCyclomaticStrict']
new_ml['SumEssential'] = ml_df['SumEssential']
new_ml['isDefective']=ml_df['isDefective'].map({False:0,True:1})
print(new_ml.info())

join_dataset = pd.concat([new_dataset,new_ml])
print(join_dataset.info())
print(join_dataset.head())

new_dataset.to_csv("ar3_apache.csv")
new_ml.to_csv("ar3_kmean.csv")
join_dataset.to_csv('apche_zing.csv')
#file = open("matrix_ml_zing.txt", "w")
#file.write(str(join_dataset))
#file.close()
```
