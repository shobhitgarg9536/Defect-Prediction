```python
import pandas as pd
```

```python
ml_df = pd.read_csv("ML.csv")
zxing_df = pd.read_csv("zxing.csv")
ml_df = ml_df[['LDHH_cbo','LDHH_fanIn','WCHU_dit','WCHU_fanOut',
             'WCHU_lcom','WCHU_numberOfAttributesInherited','WCHU_numberOfMethods','WCHU_numberOfMethodsInherited',
             'ck_oo_dit','ck_oo_fanOut','numberOfCriticalBugsFoundUntil:','LDHH_numberOfMethodsInherited',
             'LDHH_fanOut','LDHH_cbo','LDHH_numberOfMethods','class']]

zxing_df = zxing_df[['AvgLineCode','AvgLineComment','CountLine','CountLineBlank','CountLineCode','CountLineCodeDecl',
                   'CountLineCodeExe','CountLineComment','CountSemicolon','CountStmtDecl','MaxCyclomatic','MaxCyclomaticModified',
                   'SumCyclomaticModified','SumCyclomaticStrict','CountStmtExe','isDefective']]

#print(ml_df.info())
new_dataset = pd.DataFrame()
new_dataset['LDHH_numberOfPublicMethods'] = zxing_df['AvgLineComment']
new_dataset['CvsLogEntropy'] = zxing_df['AvgLineComment']
new_dataset['LDHH_numberOfPrivateAttributes'] = zxing_df['CountLineCode']
new_dataset['LDHH_lcom'] = zxing_df['CountLineCodeDecl']
new_dataset['LDHH_numberOfMethods'] = zxing_df['CountLineCodeExe']
new_dataset['LDHH_cbo'] = zxing_df['CountLineComment']
new_dataset['LDHH_fanOut'] = zxing_df['CountSemicolon']
new_dataset['LDHH_numberOfAttributes'] = zxing_df['MaxCyclomatic']
new_dataset['CvsLinEntropy'] = zxing_df['SumCyclomaticModified']
new_dataset['LDHH_numberOfPublicAttributes'] = zxing_df['SumCyclomaticStrict']
new_dataset['LDHH_numberOfAttributesInherited'] = zxing_df['SumCyclomaticStrict']
new_dataset['defects'] = zxing_df['isDefective'].map({False:1,True:0})
print(new_dataset.info())

new_ml = pd.DataFrame()
new_ml['LDHH_numberOfPublicMethods'] = ml_df['LDHH_numberOfPublicMethods']
new_ml['CvsLogEntropy'] = ml_df['CvsLogEntropy']
new_ml['LDHH_numberOfPrivateAttributes'] = ml_df['LDHH_numberOfPrivateAttributes']
new_ml['LDHH_lcom'] = ml_df['LDHH_lcom']
new_ml['LDHH_numberOfMethods'] = ml_df['LDHH_numberOfMethods']
new_ml['LDHH_cbo'] = ml_df['LDHH_cbo']
new_ml['LDHH_fanOut'] = ml_df['LDHH_fanOut']
new_ml['LDHH_numberOfAttributes'] = ml_df['LDHH_numberOfAttributes']
new_ml['CvsLinEntropy'] = ml_df['CvsLinEntropy']
new_ml['LDHH_numberOfAttributesInherited'] = ml_df['LDHH_numberOfAttributesInherited']
new_ml['LDHH_numberOfPublicAttributes'] = ml_df['LDHH_numberOfPublicAttributes']
new_ml['defects']=ml_df['class'].map({'clean':0,'buggy':1})
print(new_ml.info())

join_dataset = pd.concat([new_dataset,new_ml])
print(join_dataset.info())
print(join_dataset.head())

new_dataset.to_csv("zxing_kmean.csv")
new_ml.to_csv("ml_kmean.csv")
join_dataset.to_csv('matrix_ml_zing.csv')
#file = open("matrix_ml_zing.txt", "w")
#file.write(str(join_dataset))
#file.close()
```
