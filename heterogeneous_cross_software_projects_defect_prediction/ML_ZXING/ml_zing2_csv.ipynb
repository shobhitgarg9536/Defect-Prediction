```python
import pandas as pd
```

```python
ml_df = pd.read_csv("ML.csv")
zxing_df = pd.read_csv("zxing.csv")
ml_df = ml_df[['LDHH_cbo','LDHH_fanIn','WCHU_dit','WCHU_fanOut',
             'WCHU_lcom','WCHU_numberOfAttributesInherited','WCHU_numberOfMethods','WCHU_numberOfMethodsInherited',"WCHU_wmc",
             'ck_oo_dit','ck_oo_fanOut','numberOfCriticalBugsFoundUntil:','numberOfHighPriorityBugsFoundUntil:','class']]

zxing_df = zxing_df[['CountStmt','SumCyclomaticModified','CountStmtExe','SumCyclomatic','isDefective']]

#print(ml_df.info())
new_dataset = pd.DataFrame()
new_dataset['LDHH_cbo'] = zxing_df['CountStmt']
new_dataset['LDHH_fanIn'] = zxing_df['CountStmt']
new_dataset['WCHU_dit'] = zxing_df['SumCyclomaticModified']
new_dataset['WCHU_fanOut'] = zxing_df['CountStmt']
new_dataset['WCHU_lcom'] = zxing_df['CountStmtExe']
new_dataset['WCHU_numberOfAttributesInherited'] = zxing_df['SumCyclomaticModified']
new_dataset['WCHU_numberOfMethods'] = zxing_df['CountStmt']
new_dataset['WCHU_numberOfMethodsInherited'] = zxing_df['CountStmt']
new_dataset['WCHU_wmc'] = zxing_df['CountStmt']
new_dataset['ck_oo_dit'] = zxing_df['CountStmt']
new_dataset['ck_oo_fanOut'] = zxing_df['SumCyclomatic']
new_dataset['numberOfCriticalBugsFoundUntil:'] = zxing_df['SumCyclomaticModified']
new_dataset['numberOfHighPriorityBugsFoundUntil:'] = zxing_df['CountStmt']
new_dataset['defects'] = zxing_df['isDefective'].map({False:0,True:1})
print(new_dataset.info())

new_ml = pd.DataFrame()
new_ml['LDHH_cbo'] = ml_df['LDHH_cbo']
new_ml['LDHH_fanIn'] = ml_df['LDHH_fanIn']
new_ml['WCHU_dit'] = ml_df['WCHU_dit']
new_ml['WCHU_fanOut'] = ml_df['WCHU_fanOut']
new_ml['WCHU_lcom'] = ml_df['WCHU_lcom']
new_ml['WCHU_numberOfAttributesInherited'] = ml_df['WCHU_numberOfAttributesInherited']
new_ml['WCHU_numberOfMethods'] = ml_df['WCHU_numberOfMethods']
new_ml['WCHU_numberOfMethodsInherited'] = ml_df['WCHU_numberOfMethodsInherited']
new_ml['WCHU_wmc'] = ml_df['WCHU_wmc']
new_ml['ck_oo_dit'] = ml_df['ck_oo_dit']
new_ml['ck_oo_fanOut'] = ml_df['ck_oo_fanOut']
new_ml['numberOfCriticalBugsFoundUntil:'] = ml_df['numberOfCriticalBugsFoundUntil:']
new_ml['numberOfHighPriorityBugsFoundUntil:'] = ml_df['numberOfHighPriorityBugsFoundUntil:']
new_ml['defects']=ml_df['class'].map({'clean':0,'buggy':1})
print(new_ml.info())

join_dataset = pd.concat([new_dataset,new_ml])
print(join_dataset.info())
print(join_dataset.head())

new_dataset.to_csv("zxing_kmean2.csv")
new_ml.to_csv("ml_kmean2.csv")
join_dataset.to_csv('matrix_ml_zing2.csv')
#file = open("matrix_ml_zing.txt", "w")
#file.write(str(join_dataset))
#file.close()
```
