```python
import pandas as pd
```

```python
ml_df = pd.read_csv("ar6.csv")
zxing_df = pd.read_csv("pde.csv")
ml_df = ml_df[['blank_loc','branch_count','code_and_comment_loc','condition_count',
             'decision_count','executable_loc','halstead_effort','halstead_error',
             'halstead_length','halstead_level','halstead_time','halstead_vocabulary',
             'halstead_volume','total_operators','unique_operands','defects']]

zxing_df = zxing_df[['WCHU_numberOfMethods','WCHU_wmc','WCHU_dit','LDHH_fanIn','WCHU_numberOfAttributesInherited','ck_oo_cbo',
                   'WCHU_numberOfMethodsInherited','numberOfNonTrivialBugsFoundUntil:','ck_oo_fanOut','class']]

#print(ml_df.info())
new_dataset = pd.DataFrame()
new_dataset['blank_loc'] = zxing_df['WCHU_numberOfMethods']
new_dataset['branch_count'] = zxing_df['WCHU_wmc']
new_dataset['code_and_comment_loc'] = zxing_df['WCHU_dit']
new_dataset['condition_count'] = zxing_df['LDHH_fanIn']
new_dataset['decision_count'] = zxing_df['WCHU_numberOfAttributesInherited']
new_dataset['executable_loc'] = zxing_df['ck_oo_cbo']
new_dataset['halstead_effort'] = zxing_df['WCHU_numberOfAttributesInherited']
new_dataset['halstead_error'] = zxing_df['LDHH_fanIn']
new_dataset['halstead_length'] = zxing_df['WCHU_numberOfMethodsInherited']
new_dataset['halstead_level'] = zxing_df['numberOfNonTrivialBugsFoundUntil:']
new_dataset['halstead_time'] = zxing_df['WCHU_numberOfMethods']
new_dataset['halstead_vocabulary'] = zxing_df['WCHU_numberOfAttributesInherited']
new_dataset['halstead_volume'] = zxing_df['LDHH_fanIn']
new_dataset['total_operators'] = zxing_df['ck_oo_cbo']
new_dataset['unique_operands'] = zxing_df['ck_oo_fanOut']
new_dataset['class'] = zxing_df['class'].map({'clean':0,'buggy':1})
print(new_dataset.info())

new_ml = pd.DataFrame()
new_ml['blank_loc'] = ml_df['blank_loc']
new_ml['branch_count'] = ml_df['branch_count']
new_ml['code_and_comment_loc'] = ml_df['code_and_comment_loc']
new_ml['condition_count'] = ml_df['condition_count']
new_ml['decision_count'] = ml_df['decision_count']
new_ml['executable_loc'] = ml_df['executable_loc']
new_ml['halstead_effort'] = ml_df['halstead_effort']
new_ml['halstead_error'] = ml_df['halstead_error']
new_ml['halstead_length'] = ml_df['halstead_length']
new_ml['halstead_level'] = ml_df['halstead_level']
new_ml['halstead_time'] = ml_df['halstead_time']
new_ml['halstead_vocabulary'] = ml_df['halstead_vocabulary']
new_ml['halstead_volume'] = ml_df['halstead_volume']
new_ml['total_operators'] = ml_df['total_operators']
new_ml['unique_operands'] = ml_df['unique_operands']
new_ml['defects']=ml_df['defects'].map({False:1,True:0})
print(new_ml.info())

join_dataset = pd.concat([new_dataset,new_ml])
print(join_dataset.info())
print(join_dataset.head())

new_dataset.to_csv("ar6_pde.csv")
new_ml.to_csv("pde_kmean.csv")
join_dataset.to_csv('ar6_zing.csv')
#file = open("matrix_ml_zing.txt", "w")
#file.write(str(join_dataset))
#file.close()
```
